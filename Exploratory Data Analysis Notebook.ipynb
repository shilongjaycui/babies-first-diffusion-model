```python
import numpy as np
import matplotlib.pyplot as plt
from pickle import dump,load
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
batch_1 = unpickle('cifar-10-batches-py/data_batch_1')

batch = dict()
for key in batch_1.keys():
    batch[str(key)[2:-1]] = batch_1[key]
```

```python
batch.keys()
```

```python
labels = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
for idx in range(10):
    r = batch['data'][idx][:1024].reshape((32,32))
    g = batch['data'][idx][1024:1024*2].reshape((32,32))
    b = batch['data'][idx][1024*2:1024*3].reshape((32,32))

    image = np.zeros((32,32,3))
    image[:,:,0] = r
    image[:,:,1] = g
    image[:,:,2] = b

    plt.title("This is a " +  str(labels[batch['labels'][idx]]))
    plt.imshow(image/255)
    plt.show()
```

```python
σ = np.linspace(0.00001,1,100)
for i in range(100):
    r = np.array(batch['data'][idx][:1024].reshape((32,32))).astype(np.float32)
    g = np.array(batch['data'][idx][1024:1024*2].reshape((32,32))).astype(np.float32)
    b = np.array(batch['data'][idx][1024*2:1024*3].reshape((32,32))).astype(np.float32)

    r += np.random.normal(0,σ[i]*127.5,(32,32))
    g += np.random.normal(0,σ[i]*127.5,(32,32))
    b += np.random.normal(0,σ[i]*127.5,(32,32))

    r /= 255
    g /= 255
    b /= 255

    
    image = np.zeros((32,32,3))
    image[:,:,0] = r/np.max(r)
    image[:,:,1] = g/np.max(g)
    image[:,:,2] = b/np.max(b)

    plt.title("This is a Noisey Image")
    plt.imshow(image)
    plt.show()
```
